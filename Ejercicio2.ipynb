```python
import numpy as np
import matplotlib.pyplot as plt

y = -2
lhs = (y / 2) ** 2 

sol1 = -3 + 2 * np.sqrt(2)
sol2 = -3 - 2 * np.sqrt(2)

print("Puntos de intersecci√≥n con y = -2:")
print(f"({sol1:.4f}, -2) y ({sol2:.4f}, -2)")

x = np.linspace(-8, 2, 400)
rhs = ((x + 3) / 2) ** 2 - 1
valid = rhs >= 0 
Y_pos = 2 * np.sqrt(rhs[valid])
Y_neg = -Y_pos

plt.plot(x[valid], Y_pos, label='Curva positiva')
plt.plot(x[valid], Y_neg, label='Curva negativa')
plt.axhline(-2, color='red', linestyle='--', label='y = -2')
plt.scatter([sol1, sol2], [-2, -2], color='black', zorder=5)
plt.legend()
plt.title("Intersecci√≥n de la curva con y = -2")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Coordenadas de intersecci√≥n con y = -2
y = -2
lhs = (y / 2) ** 2 
sol1 = -3 + 2 * np.sqrt(2)
sol2 = -3 - 2 * np.sqrt(2)

# Dominio y valores de la curva
x = np.linspace(-8, 2, 400)
rhs = ((x + 3) / 2) ** 2 - 1
valid = rhs >= 0
Y_pos = 2 * np.sqrt(rhs[valid])
Y_neg = -Y_pos

# Configuraci√≥n inicial de la figura
fig, ax = plt.subplots()
ax.plot(x[valid], Y_pos, label='Curva positiva')
ax.plot(x[valid], Y_neg, label='Curva negativa')
ax.axhline(-2, color='red', linestyle='--', label='y = -2')
ax.scatter([sol1, sol2], [-2, -2], color='black', zorder=5, label='Puntos de intersecci√≥n')
point, = ax.plot([], [], 'bo', markersize=8, label='Punto en movimiento')

ax.set_xlim(-8, 2)
ax.set_ylim(-6, 6)
ax.set_title("Animaci√≥n del punto hacia la intersecci√≥n con y = -2")
ax.grid()
ax.legend()

# Trayectoria del punto
start = np.array([-7, 5])   # Punto inicial
end = np.array([sol1, -2])  # Punto de intersecci√≥n
steps = 100
trajectory = np.linspace(start, end, steps)

# Funci√≥n de animaci√≥n
def update(frame):
    point.set_data([trajectory[frame][0]], [trajectory[frame][1]])
    return point,


ani = FuncAnimation(fig, update, frames=steps, interval=50, blit=True)

# Guardar como GIF
ani.save("Ejercicio2.gif", writer=PillowWriter(fps=20))
plt.show()
```
