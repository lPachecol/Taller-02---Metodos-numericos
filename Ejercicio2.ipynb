```python
import numpy as np
import matplotlib.pyplot as plt

y = -2
lhs = (y**2)**2 

sol1 = -32 + np.sqrt(17)
sol2 = -32 - np.sqrt(17)
print("Puntos de intersecci√≥n con y = -2:")
print(f"({sol1:.4f}, -2) y ({sol2:.4f}, -2)")

x = np.linspace(-50, -10, 400)
X = []
Y = []
Y_neg = []

for xi in x:
    radicando = (xi + 32)**2 - 1
    if radicando >= 0:
        y_val = np.sqrt(np.sqrt(radicando))
        X.append(xi)
        Y.append(y_val)
        Y_neg.append(-y_val)

plt.plot(X, Y, label='Curva positiva')
plt.plot(X, Y_neg, label='Curva negativa')
plt.axhline(-2, color='red', linestyle='--', label='y = -2')
plt.scatter([sol1, sol2], [-2, -2], color='black', zorder=5)
plt.legend()
plt.title("Intersecci√≥n de la curva con y = -2")
plt.grid()
plt.show()
```
