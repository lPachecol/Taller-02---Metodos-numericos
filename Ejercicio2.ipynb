```python
import numpy as np
import matplotlib.pyplot as plt

y = -2
lhs = (y / 2) ** 2 

sol1 = -3 + 2 * np.sqrt(2)
sol2 = -3 - 2 * np.sqrt(2)

print("Puntos de intersecci√≥n con y = -2:")
print(f"({sol1:.4f}, -2) y ({sol2:.4f}, -2)")

x = np.linspace(-8, 2, 400)
rhs = ((x + 3) / 2) ** 2 - 1
valid = rhs >= 0 
Y_pos = 2 * np.sqrt(rhs[valid])
Y_neg = -Y_pos

plt.plot(x[valid], Y_pos, label='Curva positiva')
plt.plot(x[valid], Y_neg, label='Curva negativa')
plt.axhline(-2, color='red', linestyle='--', label='y = -2')
plt.scatter([sol1, sol2], [-2, -2], color='black', zorder=5)
plt.legend()
plt.title("Intersecci√≥n de la curva con y = -2")
plt.grid()
plt.show()
```
